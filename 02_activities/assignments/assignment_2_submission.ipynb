# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, QuantileTransformer

# Preproc 1: scaled variables
preproc1 = ColumnTransformer([
    ('num', StandardScaler(), num_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)])
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
# Preproc 2: Non-linear Transformation
preproc2 = ColumnTransformer([
    ('num', QuantileTransformer(output_distribution='normal'), num_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
])
```

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error

# Defining base models
baseline = Ridge()
advanced = RandomForestRegressor(random_state=42)
```

```python
# Pipeline A = preproc1 + baseline
pipeline_A = Pipeline([('preprocessing', preproc1), ('regressor', baseline)])
```

```python
# Pipeline B = preproc2 + baseline
pipeline_B = Pipeline([('preprocessing', preproc2), ('regressor', baseline)])
```

```python
# Pipeline C = preproc1 + advanced model
pipeline_C = Pipeline([('preprocessing', preproc1), ('regressor', advanced)])
```

```python
# Pipeline D = preproc2 + advanced model
pipeline_D = Pipeline([('preprocessing', preproc2), ('regressor', advanced)])
```

```python
# Ensuring fair compariso via grid search on each pipeline to find the optimal hyperparameters.
# Parameters for Ridge
param_ridge = {'regressor__alpha': [0.1, 1.0, 10.0, 100.0]}

# Parameters for Random Forest
param_rf = {
    'regressor__n_estimators': [100, 200],
    'regressor__max_depth': [None, 5, 10],
    'regressor__min_samples_split': [2, 5]
}

# Grid Search Objects
grid_a = GridSearchCV(pipeline_A, param_ridge, cv=5, scoring='neg_root_mean_squared_error')
grid_b = GridSearchCV(pipeline_B, param_ridge, cv=5, scoring='neg_root_mean_squared_error')
grid_c = GridSearchCV(pipeline_C, param_rf, cv=5, scoring='neg_root_mean_squared_error')
grid_d = GridSearchCV(pipeline_D, param_rf, cv=5, scoring='neg_root_mean_squared_error')

# Fit models
grids = [grid_a, grid_b, grid_c, grid_d]
for g in grids:
    g.fit(X, y)
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.

```python
from sklearn.model_selection import GridSearchCV
import joblib
```

```python
# Parameter Grids
# Ridge: 4 values for Alpha
param_grid_ridge = {'regressor__alpha': [0.1, 1.0, 10.0, 100.0]}
```

```python
# Random Forest: 4 combinations (2 depths * 2 estimator counts)
param_grid_rf = {
    'regressor__max_depth': [5, 10],
    'regressor__n_estimators': [50, 150]
}
```

```python
# Grid Search setup
grids = {
    "Pipeline_A": GridSearchCV(pipeline_A, param_grid_ridge, cv=5, scoring='neg_mean_absolute_error'),
    "Pipeline_B": GridSearchCV(pipeline_B, param_grid_ridge, cv=5, scoring='neg_mean_absolute_error'),
    "Pipeline_C": GridSearchCV(pipeline_C, param_grid_rf, cv=5, scoring='neg_mean_absolute_error'),
    "Pipeline_D": GridSearchCV(pipeline_D, param_grid_rf, cv=5, scoring='neg_mean_absolute_error')
}
```

# Evaluate

+ Which model has the best performance?

```python
best_score = float('-inf')
best_model = None

for name, grid in grids.items():
    grid.fit(X, y)
    print(f"{name} Best MAE: {-grid.best_score_:.4f}")
    if grid.best_score_ > best_score:
        best_score = grid.best_score_
        best_model = grid.best_estimator_
```
